Here we can see that there are some duplicate data that we are going to eliminate, and we are also going to transform the parking column into yes or no values.

```python
df = df.drop_duplicates()
```

```python
df['Parking'] = df['Parking'].apply(lambda x: 'Yes' if pd.notna(x) and x.strip() != "" else 'No')
```

```python
df = df.drop(columns=['Transaction_Number', 'Transaction_Date', 'Transaction_sub_type', 'Is_Free_Hold?', 'Usage', 'Transaction_Size', 'Room(s)', 'No_of_Buyer', 'No_of_Seller', 'Master_Project', 'Latitude_Project', 'Longitude_Project'], axis=1)
```
